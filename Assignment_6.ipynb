**4.** Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
z=np.cos(x)
plt.plot(x,y,':b',label='sine',color="royalblue",linewidth=3)
plt.plot(x,z,'g--',label='cosine',color="limegreen",linewidth=3)
plt.xlabel('Angle',weight = 'bold')
plt.ylabel('Magnitude',weight = 'bold')
plt.title('Sine_cosine_curve',fontweight = 'bold')
plt.legend()
plt.show()
```
